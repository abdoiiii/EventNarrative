```python
import os
import numpy as np
import matplotlib.pyplot as plt
import textwrap
f = lambda x: textwrap.fill(x.get_text(), 20)
relations = ['winner', 'sports season of league or competition', 'part of', 'end time', 'instance of', 'start time', 'sport', 'country', 'location', 'point in time']
y_pos = np.arange(len(relations))
values = [4.08, 4.27, 4.38, 4.58, 4.66, 4.67, 4.71, 4.80, 4.99, 5.06]

relations.reverse()
values.reverse()
fig, ax = plt.subplots()
hbars = ax.barh(y_pos, values, height=0.5,align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(relations)
ax.set_yticklabels(map(f, ax.get_yticklabels()))
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('log(count)',size=14 )
ax.set_yticklabels(relations, rotation=15)
ax.set_ylabel('Relations',size=14)
ax.set_title('Top 10 Relations',size=18)

# Label with specially formatted floats
ax.bar_label(hbars, fmt='%.2f')
ax.set_xlim(right=6)  # adjust xlim to fit labels
plt.savefig('relations.pdf', transparent=True,bbox_inches="tight")
# plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import textwrap
f = lambda x: textwrap.fill(x.get_text(), 20)
types = ['sports season', 'American football team season','sporting event','tennis tournament edition','battle','election','basketball team season','association football team season','Olympic sporting event','nation at sport competition']
y_pos = np.arange(len(types))
values = [4.96,4.49,4.43,4.15,4.14,4.08,4.08,3.99,3.95,3.88]
fig, ax = plt.subplots()
hbars = ax.barh(y_pos, values, height=0.5,align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(types)
ax.set_yticklabels(map(f, ax.get_yticklabels()))
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('log(count)',size=14 )
ax.set_yticklabels(types, rotation=15)
ax.set_ylabel('Types',size=14)
ax.set_title('Top 10 Types',size=18)
ax.bar_label(hbars, fmt='%.2f')
ax.set_xlim(right=6)  # adjust xlim to fit labels
plt.savefig('types.pdf', transparent=True,bbox_inches="tight")
```
